### 네이버 연관검색어 데이터 수집
- 정적페이지 : html(str) : BeautifulSoup : bs_obj.select(css-selecter) : text

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 1. 웹서비스 분석 : url
```

```python
keyword = '삼성전자'
url = f'https://search.naver.com/search.naver?query={keyword}'
```

```python
# 2. request(url) > response(data) : data(html)
# html문자열은 한페이지 모두
```

```python
response = requests.get(url)
response
```

```python
response.text
```

```python
# 3. data(html) > bs_obj.sleect(css-selecter) > text
```

```python
dom = BeautifulSoup(response.text, "html.parser")
```

```python
type(dom)
```

```python
# select() : 여러개의 엘리먼트 선택 : list : [Tag, Tag, Tag ...]
# select_one() : 하나의 엘리먼트 선택 : Tag
```

```python
# li 엘리먼트 10개 선택
```

```python
elements = dom.select('.lst_related_srch > li')
```

```python
len(elements), type(elements[0])
```

```python
# 반복문으로 각각의 엘리먼트에서 text 추출
```

```python
element = elements[0]
element.select_one('.tit').text
```

```python
relational_keyword = [element.select_one('.tit').text for element in elements]
relational_keyword
```

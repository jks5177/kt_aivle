# 실습 내용

- Grid Search로 다양한 알고리즘 모델을 튜닝합니다.


# 1.환경 준비

- 기본 라이브러리와 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/jangrae/csv/master/diabetes.csv'
data = pd.read_csv(path)
```

# 2.데이터 이해

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

```python
# 상위 몇 개 행 확인
data.head()
```

```python
# 변수 확인
data.info()
```

```python
# 기초통계량 확인
data.describe()
```

```python
# NaN 값 확인
data.isnull().sum()
```

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) x, y 분리**

```python
# target 확인
target = 'Outcome'

# 데이터 분리
x = data.drop(target, axis = 1)
y = data[target] 
```

**2) 정규화**

```python
# 정규화
x = (x - x.min()) / (x.max() - x.min())
```

**3) 학습용, 평가용 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2022)
```

# 4.모델링

- GridSearchCV를 사용해 최선의 하이퍼파라미터를 갖는 모델을 만듭니다.
- 우선 사용할 모듈을 모두 불러옵니다.
- 평가 지표는 Classification Report를 사용합니다.

```python
# 모듈 불러오기
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import classification_report
```

## 4.1.KNN

- 파라미터 
    - n_neighbors: range(1, 11)
- GridSearchCV 선언
    - cv = 5
    - scoring = 'accuracy'

```python
# 선언하기
model_knn = KNeighborsClassifier()

# 파라미터 선언
param = {'n_neighbors' : range(1, 11)}

# Grid Search 모델 선언하기
model = GridSearchCV(model_knn,
                     param,
                     cv=5,
                     scoring='accuracy')
```

```python
# 학습하기
model.fit(x_train, y_train)
```

```python
# 최적파라미터, 성능 확인
print(model.best_params_)
print(model.best_score_)
```

```python
# 예측하기
y_pred = model.predict(x_test)
```

```python
# 평가하기
print(classification_report(y_test, y_pred))
```

```python
# 성능 수집
result = {}
result['KNN'] = classification_report(y_test, y_pred)
```

## 4.2.Decision Tree

- random_state = 2022
- 파라미터
    - max_depth: range(1, 11)
- GridSearchCV 선언
    - cv = 5
    - scoring = 'accuracy'

```python
# 선언하기
model_tree = DecisionTreeClassifier(random_state=2022)

# 파라미터 선언
param = {'max_depth':range(1,11)}

# Grid Search 모델 선언하기
model = GridSearchCV(model_tree,
                     param,
                     cv=5,
                     scoring='accuracy')
```

```python
# 학습하기
model.fit(x_train, y_train)
```

```python
# 최적파라미터, 성능 확인
print(model.best_params_)
print(model.best_score_)
```

```python
# 예측하기
y_pred = model.predict(x_test)
```

```python
# 평가하기
print(classification_report(y_test, y_pred))
```

```python
result['DecisionTree'] = classification_report(y_test, y_pred)
```

```python
# 변수 중요도 시각화
plt.barh(y=list(x), width=model.best_estimator_.feature_importances_)
plt.show()
```

```python
# 트리 시각화
from sklearn.tree import plot_tree

plt.figure(figsize=(25, 10))

plot_tree(model.best_estimator_,
          filled=True,
          feature_names=list(x),
          class_names=['No', 'Yes'],
          fontsize=6,)

plt.show()
```

## 4.3.Logistic Regression

- GridSearchCV를 사용하지 않고 모델링합니다.

```python
# 선언하기
model = LogisticRegression()
```

```python
# 학습하기
model.fit(x_train, y_train)
```

```python
# 예측하기
y_pred = model.predict(x_test)
```

```python
# 평가하기
print(classification_report(y_test, y_pred))
```

```python
result['Logistic Regression'] = classification_report(y_test, y_pred)
```

```python
# 변수 가중치 시각화
plt.barh(y=list(x), width=model.coef_.reshape(-1))
plt.show()
```

## 4.4.SVM

- 파라미터
    - C: [0.0001, 0.001, 0.01, 1, 10, 100, 1000, 10000]
    - gamma: [0.00001, 0.0001, 0.001, 0.01, 1, 10, 100]
- GridSearchCV 선언
    - cv = 5
    - scoring = 'accuracy'

```python
# np.logspace(시작 0의 개수, 마지막 0의 개수, 숫자의 개수)
print(np.logspace(-3, 4, 8)) # [0.0001, 0.001, 0.01, 1, 10, 100, 1000, 10000]
```

```python
# 선언하기
model_svm = SVC()

# 파라미터 선언
param = {'C' : [0.0001, 0.001, 0.01, 1, 10, 100, 1000, 10000],
             'gamma' : [0.00001, 0.0001, 0.001, 0.01, 1, 10, 100]}

# Grid Search 모델 선언하기
model = GridSearchCV(model_svm,          # 기본 모델 이름
                     param,              # 앞에서 선언한 튜닝용 파라미터 변수
                     cv = 5,             # k-Fold Cross Validation(default=5)
                     scoring='accuracy') # 평가 방법
```

```python
# 학습하기
model.fit(x_train, y_train)
```

```python
# 최적파라미터, 성능 확인
print(model.best_params_)
print(model.best_score_)
```

```python
# 예측하기
y_pred = model.predict(x_test)
```

```python
# 평가하기
print(classification_report(y_test, y_pred))
```

```python
result['SVM'] = classification_report(y_test, y_pred)
```

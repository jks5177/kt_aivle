```python
def review_page(productNo) :
    url = f'https://brand.naver.com/n/v1/reviews/paged-reviews?page=1&pageSize=20&merchantNo=500030812&originProductNo={productNo}&sortType=REVIEW_RANKING'
    response = requests.get(url)
    return response.json()['totalPages']
```

```python
# review function
def review(productNp, page=1) :
    
    url = f'https://brand.naver.com/n/v1/reviews/paged-reviews?page={page}&pageSize=20&merchantNo=500030812&originProductNo={productNo}&sortType=REVIEW_RANKING'
    response = requests.get(url)
    datas = response.json()['contents']
    columns = ['id', 'reviewScore', 'createDate', 'reviewContent']
    return pd.DataFrame(datas)[columns]
```

```python
# 데이터가 너무 많아 1000 page만 수집하는 것으로 test

item = best_item_df.loc[0]
productId = item['link'].split('/')[-1]
productNo = find_productNo(productId)

totalpage = review_page(productNo)
if totalpage > 500 :
    totalpage = 500

review_datas = []
for i in range(totalpage) :
    print(i, end=' ')
    review_datas.append(review(productNp=productNo, page=i+1))
```

```python
len(review_datas)
```

```python
review_df = pd.concat(review_datas, ignore_index=True)
review_df.tail()
```

### 추가 best item 모든 리뷰 크롤링

```python
review_df.to_csv(item['item']+'.csv')
```

```python
# best item의 500 page review 크롤링
for idx, item in best_item_df.iterrows() :
    print(item['item'])
    productId = item['link'].split('/')[-1]
    productNo = find_productNo(productId = productId)

    totalpage = review_page(productNo)
    if totalpage > 500 :
        totalpage = 500

    review_datas = []
    for i in range(totalpage) :
        print(i, end=' ')
        review_datas.append(review(productNp=productNo, page=i+1))
        
    print()
    review_df = pd.concat(review_datas)
    review_df.to_csv(item['item'].replace('/', '')+'.csv')
```
